# 📊 Superstore Sales Data Analysis

## Project Objective

The objective of this project is to analyze the Superstore sales dataset to understand sales performance, profitability, customer segments, and regional trends.

The analysis aims to identify business opportunities and provide data-driven recommendations that can help improve sales and profit.

```python




import pandas as pd
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('Sample_Superstore.csv', encoding='latin-1')
df.head()
```

### Observations

- The dataset contains information about orders, customers, products, sales, and profit.
- There are no missing values.
- Sales and Profit are numerical variables.
- Order Date and Ship Date need to be converted to datetime format.

## Data Cleaning

```python
df = df.drop_duplicates()
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"]= pd.to_datetime(df["Ship Date"])
df.head()
```

### Which product category generates the highest sales?

```python
Category_Sales = df.groupby("Category")["Sales"].sum()
Category_Sales.plot(kind="bar")

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")

plt.show()
```

### Which region is the most profitable?

```python
region_profit = df.groupby("Region")["Profit"].sum()
region_profit.plot(kind="bar")
plt.title("Most profitable Regions ")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()

```

### Business Insight

The West region generates the highest total profit, while the Central region has the lowest.

This indicates that the company performs better in the West and may benefit from applying similar sales or marketing strategies in lower-performing regions.

### Do higher discounts reduce profit,?

```python
plt.figure(figsize=(8, 5))

plt.scatter(df["Discount"], df["Profit"] )
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

```python
df[["Discount", "Profit"]].corr()
```

### Findings

The analysis suggests that higher discounts are generally associated with lower profits. Many orders with large discounts generate very little profit or even losses.

### Recommendation

The company should review its discount strategy. Instead of offering large discounts on many products, it should identify products where discounts increase sales without reducing profit.

### How do sales change over time?

```python
df["Monthh"]= df["Order Date"].dt.month_name()

month_order = [
    "January", "February", "March", "April",
    "May", "June", "July", "August",
    "September", "October", "November", "December"
]

df["Monthh"] = pd.Categorical(
    df["Monthh"],
    categories=month_order,
    ordered=True
)
```

```python
monthly_sales = df.groupby("Monthh")["Sales"].sum()
monthly_sales
```

```python
plt.figure(figsize=(10,5))

monthly_sales.plot(marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)

plt.show()
```

### Findings

Sales fluctuate throughout the year. Some months show noticeably higher sales, indicating seasonal demand.



### Recommendation

Increase inventory and promotional activities before high-sales months.

During lower-sales months, consider offering targeted promotions or introducing new products to encourage customer purchases.

### Which products generate the highest sales?

```python
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_products
```

```python
plt.figure(figsize=(12,6))

top_products.plot(kind="bar")

plt.title("Top 10 Products by Sales")
plt.xlabel("Product")
plt.ylabel("Total Sales")



plt.show()
```

### Findings

The top-selling products contribute a significant portion of the company's total sales. These products are consistently popular and play an important role in generating revenue.

### Recommendations

- Ensure top-selling products are always in stock.
- Increase marketing efforts for these products.
- Monitor inventory levels to avoid stock shortages.
- Consider bundling these products with lower-selling items to increase overall sales.

### Which sub-categories are losing money?

```python
subcat_profit = df.groupby("Sub-Category")["Profit"].sum().sort_values()

subcat_profit
```

```python
plt.figure(figsize=(12,6))

subcat_profit.plot(kind="bar")

plt.title("Profit by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Total Profit")

plt.xticks(rotation=45)

plt.show()
```

### Findings

Some sub-categories generate negative or very low profit, meaning the company is losing money on these products despite making sales.

This indicates that high sales do not always guarantee profitability.

### Recommendations

- Investigate loss-making sub-categories to understand the cause (high discount, low pricing, or high cost).
- Reduce or control discounts on low-profit items.
- Consider increasing prices slightly if the market allows.
- Remove or redesign products that consistently generate losses.
- Focus marketing on profitable sub-categories instead.

### Which shipping mode is most used and most profitable?

```python
ship_usage = df["Ship Mode"].value_counts()

ship_usage
```

```python
ship_usage.plot(kind="bar")

plt.title("Most Used Shipping Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Number of Orders")

plt.show()
```

```python
ship_profit = df.groupby("Ship Mode")["Profit"].sum().sort_values(ascending=False)

ship_profit
```

```python
ship_profit.plot(kind="bar")

plt.title("Profit by Shipping Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Total Profit")

plt.show()
```

### Findings

Standard Class is the most frequently used shipping mode, shows that customers prefer cheaper or more flexible delivery options.

However, faster shipping methods such as First Class or Same Day may generate lower profit levels.

### Recommendations

- Continue offering Standard Class as it is the most popular option.
- Analyze whether faster shipping methods improve customer satisfaction enough to justify their cost.
- Optimize logistics to reduce shipping costs where possible.
- Consider pricing strategies for premium shipping options.
